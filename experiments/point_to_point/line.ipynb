```python
from agents.rtg import RTGAgent
from environment.models import MovableObject
from environment.p2p_environment import P2PEnvironment
```

# Setup

```python
obj1 = MovableObject(name="Object", pos_x=0, pos_y=0, pos_z=0)
env = P2PEnvironment(objects=[obj1])
```

```python
for i in range(env.size):
    env.step((1, 1, 1))
    reward = env._calculate_reward()
    if env.episode_finished()[0]:
        break
env.render()
```

```python
env.reset()
```

# RTG

```python
rtg = RTGAgent(obs_dim=6, hidden_sizes=[32], learning_rate=1e-3)
rtg.train(env=env)
```
